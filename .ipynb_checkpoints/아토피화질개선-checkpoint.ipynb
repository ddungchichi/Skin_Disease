```python
from selenium import webdriver
import time
import os
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import urllib.request
from urllib.error import HTTPError
```

```python
service = Service(executable_path = "chromedriver.exe")
driver = webdriver.Chrome()
driver.get("http://www.google.co.kr/imghp?h1=ko&ogb1")
elem = driver.find_element(By.CLASS_NAME, "gLFyf")
elem.send_keys("Psoriasis photo")
elem.send_keys(Keys.RETURN)

SCROLL_PAUSE_TIME = 1
# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)
    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        try:
            driver.find_elements(By.CSS_SELECTOR, ".mye4qd").click()
        except:
            break
    last_height = new_height

# 이미지 수집
images = driver.find_elements(By.CSS_SELECTOR, ".F0uyec")
print(f"이미지 개수: {len(images)}")
count = 361
start_idx = 100
check = 1
fail = 1
num = 1
# 이미지 다운로드 루프
for image in images:
    if start_idx > num:
        num += 1
        continue
    try:
        # 이미지 클릭
        driver.execute_script("arguments[0].click();", image)

        # 기본 XPATH로 이미지 대기 및 URL 추출 시도
        try:
            element = WebDriverWait(driver, 2).until(
                EC.visibility_of_element_located((By.XPATH, '//*[@id="Sva75c"]/div[2]/div[2]/div/div[2]/c-wiz/div/div[2]/div/a/img[1]')))
            # imgUrl = element.get_attribute('src')
        except Exception as e:
            # 기본 XPATH 실패 시 대체 XPATH 시도
            try:
                element = WebDriverWait(driver, 2).until(
                EC.visibility_of_element_located((By.XPATH, '//*[@id="Sva75c"]/div[2]/div[2]/div/div[2]/c-wiz/div/div[3]/div[1]/a/img[1]')))
                # imgUrl = element.get_attribute('src')
                print(f"대체 XPATH 성공 (check: {check})")
                check += 1
            except Exception as e:
                fail += 1
                continue
                
        imgUrl = element.get_attribute('src')
        # 이미지 저장 경로 설정
        folder_path = r'C:\Users\yohee\SkinDisease_AI\Psoriasis'
        download_path = os.path.join(folder_path, f"{count}.jpg")
        # 이미지 다운로드
        urllib.request.urlretrieve(imgUrl, download_path)
        print(f"다운로드 완료: {count}")
        count += 1
        num += 1

    except HTTPError as e:
        print(f"HTTPError 발생: {e}")
    except Exception as e:
        print(f"기타 오류 발생: {str(e)}")

print(f"count : {count}")
print(f"check : {check}")
print(f"fail : {fail}")
# 드라이버 종료
driver.quit()
```
